# color reference


# Marker size (ms)


```python
k = np.array([3,8,1,10])
plt.plot(k, marker= '<', ms=15);
```

# marker Edge Color (mec)


```python
k = np.array([3,8,1,10])
plt.plot(k, marker= '*', ms=15, mec='k')
plt.show()
```

# Marker Face Color (mfc)


```python
k = np.array([3,8,1,10])
plt.plot(k, marker= '*', ms=25, mec='k', mfc = 'r')
plt.show()
```

```python
k = np.array([3,8,1,10])
plt.plot(k, marker= '*', ms=25, mec='#9DEE10', mfc = '#1507FA')
plt.show()
```

## SCATTER PLOT
her gözlem değeri için bir nokta çizer. 

```python
x = np.array([5,7,8,10,20,30,2,1]) # araba yaşı
y = np.array([100,80,70,60,50,40,150,200]) # arabanın hızı 

plt.scatter(x,y)
plt.show()
```

```python
x = np.array([5,7,8,10,20,30,2,1]) # araba yaşı
y = np.array([100,80,70,60,50,40,150,200]) # arabanın hızı 

plt.scatter(x,y)


x = np.array([5,7,8,10,20,30,2,1]) # araba yaşı
y = np.array([100,80,70,60,50,40,150,500]) # arabanın hızı 
sizes= np.array([100,80,70,60,50,40,150,500])

plt.scatter(x,y,s=sizes)
plt.show()
```

```python
#fig, ax 


x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))

colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x,y, c=colors,s=sizes, alpha=0.5, cmap='nipy_spectral' )

plt.colorbar()

plt.show()
```
